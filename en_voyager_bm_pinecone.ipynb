### dense embedding and pinecone upserting

```python
import voyageai
from dotenv import load_dotenv
load_dotenv()
import os
import time

api_key_list = ['VOYAGE_AI1', 'VOYAGE_AI2' , 'VOYAGE_AI3' , 'VOYAGE_AI4'] # 50 million free tokens
vo1 = voyageai.Client(api_key= os.environ.get('VOYAGE_EN1'))
vo2 = voyageai.Client(api_key= os.environ.get('VOYAGE_EN2'))
vo3 = voyageai.Client(api_key= os.environ.get('VOYAGE_EN3'))
vo4 = voyageai.Client(api_key= os.environ.get('VOYAGE_EN4'))
vo = voyageai.Client(api_key= os.environ.get('VOYAGE_AI_Test'))
```

```python
import asyncio
from tqdm.asyncio import tqdm
import voyageai
import os
from pinecone import Pinecone


async def process_dataset(df, vo_client, _):
    batch_size = 128
    len_df = len(df)
    
    async for i in tqdm(range(0, len_df, batch_size)):
        i_end = min(i+batch_size, len_df)
        
        metadata_batch = df.iloc[i:i_end]['merged_metadata'].values.tolist()
        meta_dict = df.iloc[i:i_end]['metadata'].values.tolist()
        mergedtext_batch = df.iloc[i:i_end]['merged_text'].values.tolist()
        
        sparse_embeds = sparse_embed_model.embed_documents([text for text in metadata_batch])
        dense_embeds = list(vo_client.embed(mergedtext_batch, model="voyage-large-2-instruct", input_type="document", truncation=True).embeddings)
        product_ids = df.iloc[i:i_end]['product_id'].values.tolist()
        
        upserts = []
        for _id, sparse, dense, meta in zip(product_ids, sparse_embeds, dense_embeds, meta_dict):
            upserts.append({
                'id': _id,
                'sparse_values': sparse,
                'values': dense,
                'metadata': meta
            })
        
        await asyncio.to_thread(index.upsert, upserts)


async def main():
    # Create VoyageAI clients
    vo1 = voyageai.Client(api_key= os.environ.get('VOYAGE_EN1'))
    vo2 = voyageai.Client(api_key= os.environ.get('VOYAGE_EN2'))
    vo3 = voyageai.Client(api_key= os.environ.get('VOYAGE_EN3'))
    vo4 = voyageai.Client(api_key= os.environ.get('VOYAGE_EN4'))

    # Prepare datasets
    df_1= df.iloc[:109500,:]
    df_2 = df.iloc[109500:219000,:]
    df_3 = df.iloc[219000:328500,:]
    df_4 = df.iloc[328500:,:]


    # Create tasks for concurrent processing
    tasks = [
        process_dataset(df_1, vo1, "Dataset 1"),
        process_dataset(df_2, vo2, "Dataset 2"),
        process_dataset(df_3, vo3, "Dataset 3"),
        process_dataset(df_4, vo4, "Dataset 4")
    ]

    # Run all tasks concurrently
    await asyncio.gather(*tasks)

    # Print index stats
    print(index.describe_index_stats())

await main()
```

```python
import multiprocessing
from functools import partial
from ranx import Qrels, Run, evaluate
from concurrent.futures import ThreadPoolExecutor, as_completed
import time

def process_query(data, query_id_list , engine):
    relevant_doc_ids_l , relevant_score_l , query_l, recommended_doc_ids_l , recommended_scores_l = [],[],[],[],[]

    for query_id in query_id_list : 
        temp = data[data['query_id'] == query_id].sort_values(by='esci_label', ascending=False)
        relevant_doc_ids_l.append(list(temp['product_id'].values))
        relevant_score_l.append(list(temp['esci_label'].values))
        query_l.append(temp['query'].iloc[0])
    recommendations = engine.recommend(query_l, rerank_top_k=10, alpha=0.75 , batch= True)
    for i in range(len(recommendations)):
        recommended_doc_ids_l.append(recommendations[i]['product_id'].to_list())
        recommended_scores_l.append(recommendations[i]['scores'].to_list())
        time.sleep(0.3)
    return relevant_doc_ids_l, relevant_score_l, recommended_doc_ids_l, recommended_scores_l

def evaluate_rank_metric(query_ids, data, engine , TOP_VAL=None , batch_size = 128):
    Iter_val = TOP_VAL if isinstance(TOP_VAL, int) else len(query_ids)
    print("Queries under Search:", Iter_val)
    relevant_doc_ids_list = []
    relevant_scores_list = []
    recommended_doc_ids_list = []
    recommended_scores_list = []
    
    with ThreadPoolExecutor() as executor:
        futures = [executor.submit(process_query, data, query_ids[i:min(Iter_val , i + batch_size)] , engine ) for i in range(0 , Iter_val, batch_size)]
        for future in as_completed(futures):
            relevant_doc_ids, relevant_score, recommended_doc_ids, recommended_scores = future.result()
            relevant_doc_ids_list.extend(relevant_doc_ids)
            relevant_scores_list.extend(relevant_score)
            recommended_doc_ids_list.extend(recommended_doc_ids)
            recommended_scores_list.extend(recommended_scores)

    from ranx import Qrels, Run, evaluate
    qrels = Qrels()
    qrels.add_multi(
        q_ids=list(map(str, query_ids[:TOP_VAL])),
        doc_ids=relevant_doc_ids_list,
        scores=relevant_scores_list,
    )

    run = Run()
    run.add_multi(
        q_ids=list(map(str, query_ids[:TOP_VAL])),
        doc_ids=recommended_doc_ids_list,
        scores=recommended_scores_list,
    )

    eval_res = evaluate(qrels, run, ["hit_rate@1", "hit_rate@5", "hit_rate@10", "hits@1", "hits@5", "hits@10", 
                                     "precision@1", "precision@5", "precision@10", "recall@1", "recall@5", 
                                     "recall@10", "f1@1", "f1@5", "f1@10", "mrr"])

    return relevant_doc_ids_list, relevant_scores_list , recommended_doc_ids_list , recommended_scores_list ,eval_res 
    
```
