```python
import asyncio
from tqdm.asyncio import tqdm
import os
from pinecone import Pinecone


async def process_dataset(df, _):
    batch_size = 128
    len_df = len(df)
    
    async for i in tqdm(range(0, len_df, batch_size)):
        i_end = min(i+batch_size, len_df)
        
        metadata_batch = df.iloc[i:i_end]['merged_metadata'].values.tolist()
        meta_dict = df.iloc[i:i_end]['metadata'].values.tolist()
        mergedtext_batch = df.iloc[i:i_end]['merged_text'].values.tolist()
        
        sparse_embeds = sparse_embed_model.embed_documents([text for text in metadata_batch])
        dense_embeds = dense_embed_model.embed_documents(mergedtext_batch)
        product_ids = df.iloc[i:i_end]['product_id'].values.tolist()
        
        upserts = []
        for _id, sparse, dense, meta in zip(product_ids, sparse_embeds, dense_embeds, meta_dict):
            upserts.append({
                'id': _id,
                'sparse_values': sparse,
                'values': dense,
                'metadata': meta
            })
        
        await asyncio.to_thread(index.upsert, upserts)


async def main():

    # Prepare datasets
    df_1= df.iloc[:109500,:]
    df_2 = df.iloc[109500:219000,:]
    df_3 = df.iloc[219000:328500,:]
    df_4 = df.iloc[328500:,:]


    # Create tasks for concurrent processing
    tasks = [
        process_dataset(df_1, "Dataset 1"),
        process_dataset(df_2, "Dataset 2"),
        process_dataset(df_3, "Dataset 3"),
        process_dataset(df_4, "Dataset 4")
    ]

    # Run all tasks concurrently
    await asyncio.gather(*tasks)

    # Print index stats
    print(index.describe_index_stats())

await main()
```

```python
ES_df_queries_e = df_queries_e[df_queries_e['esci_label'].isin(['E','S'])]
ES_df_queries_h = df_queries_h[df_queries_h['esci_label'].isin(['E','S'])]
# easy_queries = df_queries_e['query'].unique()
# hard_queries = df_queries_h['query'].unique()
easy_queries_id = df_queries_e['query_id'].unique().tolist()
hard_queries_id = df_queries_h['query_id'].unique().tolist()
# len(easy_queries) , len(hard_queries) , 
len(easy_queries_id) , len(hard_queries_id)
ES_df_queries_e['esci_label'] = ES_df_queries_e['esci_label'].replace({'E': 2, 'S': 1})
ES_df_queries_h['esci_label'] = ES_df_queries_h['esci_label'].replace({'E': 2, 'S': 1})
```

```python
import multiprocessing
from functools import partial
from ranx import Qrels, Run, evaluate
from concurrent.futures import ThreadPoolExecutor, as_completed
import time

def process_query(data, query_id , engine):
    temp = data[data['query_id'] == query_id].sort_values(by='esci_label', ascending=False)
    relevant_doc_ids = temp['product_id'].to_list()
    relevant_score = temp['esci_label'].to_list()
    query = temp['query'].iloc[0]
    recommendations = engine.recommend(query, rerank_top_k=10, alpha=0.75)
    recommended_doc_ids = recommendations['product_id'].to_list()
    recommended_scores = recommendations['scores'].to_list()
    return relevant_doc_ids, relevant_score, recommended_doc_ids, recommended_scores

def evaluate_rank_metric(query_ids, data, engine , TOP_VAL=None):
    Iter_val = TOP_VAL if isinstance(TOP_VAL, int) else len(query_ids)
    print("Queries under Search:", Iter_val)
    relevant_doc_ids_list = []
    relevant_scores_list = []
    recommended_doc_ids_list = []
    recommended_scores_list = []


    with ThreadPoolExecutor() as executor:
        futures = [executor.submit(process_query, data, query_ids[i] , engine ) for i in range(Iter_val)]
        for future in as_completed(futures):
            relevant_doc_ids, relevant_score, recommended_doc_ids, recommended_scores = future.result()
            relevant_doc_ids_list.append(relevant_doc_ids)
            relevant_scores_list.append(relevant_score)
            recommended_doc_ids_list.append(recommended_doc_ids)
            recommended_scores_list.append(recommended_scores)

    from ranx import Qrels, Run, evaluate
    qrels = Qrels()
    qrels.add_multi(
        q_ids=list(map(str, query_ids[:TOP_VAL])),
        doc_ids=relevant_doc_ids_list,
        scores=relevant_scores_list,
    )

    run = Run()
    run.add_multi(
        q_ids=list(map(str, query_ids[:TOP_VAL])),
        doc_ids=recommended_doc_ids_list,
        scores=recommended_scores_list,
    )

    eval_res = evaluate(qrels, run, ["hit_rate@1", "hit_rate@5", "hit_rate@10", "hits@1", "hits@5", "hits@10", 
                                     "precision@1", "precision@5", "precision@10", "recall@1", "recall@5", 
                                     "recall@10", "f1@1", "f1@5", "f1@10", "mrr"])

    return relevant_doc_ids_list, relevant_scores_list , recommended_doc_ids_list , recommended_scores_list ,eval_res 
    
```
