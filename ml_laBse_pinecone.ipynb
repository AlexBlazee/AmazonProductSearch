### dense vectors and upserts


```python
from embedder.dense.laBSEMultiLingualEmbedding import LaBSEMultiLingualEmbedding
dense_embed_model = LaBSEMultiLingualEmbedding()
```

```python
index.describe_index_stats()
```

```python
import asyncio
from tqdm.asyncio import tqdm
import os
from pinecone import Pinecone


async def process_dataset(df, _):
    batch_size = 128
    len_df = len(df)
    
    async for i in tqdm(range(0, len_df, batch_size)):
        i_end = min(i+batch_size, len_df)
        
        metadata_batch = df.iloc[i:i_end]['merged_metadata'].values.tolist()
        meta_dict = df.iloc[i:i_end]['metadata'].values.tolist()
        mergedtext_batch = df.iloc[i:i_end]['merged_text'].values.tolist()
        
        sparse_embeds = sparse_embed_model.embed_documents([text for text in metadata_batch])
        dense_embeds = dense_embed_model.embed_documents(mergedtext_batch)
        product_ids = df.iloc[i:i_end]['product_id'].values.tolist()
        
        upserts = []
        for _id, sparse, dense, meta in zip(product_ids, sparse_embeds, dense_embeds, meta_dict):
            upserts.append({
                'id': _id,
                'sparse_values': sparse,
                'values': dense,
                'metadata': meta
            })
        
        await asyncio.to_thread(index.upsert, upserts)


async def main():

    # Prepare datasets
    df_1= df.iloc[:106000,:] #df_1= df.iloc[:106000,:] Break
    df_2 = df.iloc[106000:212000,:]  #df_2 = df.iloc[106000:212000,:] 
    df_3 = df.iloc[212000:318000,:] # df_3 = df.iloc[212000:318000,:] 242976
    df_4 = df.iloc[318000:,:] # df_4 = df.iloc[318000:,:] 
    # notes : embedding model broke near iter : 242 ,manually identified positions


    # Create tasks for concurrent processing
    tasks = [
        process_dataset(df_1, "Dataset 1"),
        process_dataset(df_2, "Dataset 2"),
        process_dataset(df_3, "Dataset 3"),
        process_dataset(df_4, "Dataset 4")
    ]

    # Run all tasks concurrently
    await asyncio.gather(*tasks)

    # Print index stats
    print(index.describe_index_stats())

await main()
```

### Hybrid Query Setup

```python
import os
import pickle
from pinecone import Pinecone
import voyageai
import pandas as pd
from tabulate import tabulate
# from not_used_embedding import EmbeddingModel
from embedder.hybridEmbeddingModel import HybridEmbeddingModel
from reranker.jinaAIMultiLingualReranker import JinaAIMultiLingualReranker
from dotenv import load_dotenv
from embedder.sparse.bmEmbedding import BmEmbedding
from embedder.dense.laBSEMultiLingualEmbedding import LaBSEMultiLingualEmbedding
from RecommendationEngine import RecommendationEngine

load_dotenv()
# pinecone VD index name
# index_name = os.environ.get('PINECONE_INDEX_NAME')
index_name = "product-search-labse-langchain-pinecone"
print(f"Index name is {index_name}")
# pinecone -api -key 
pinecone_apikey = os.environ.get('PINECONE_LABSE_API_KEY') 
# setup sparse encoding model
sparse_embed_model = BmEmbedding()
sparse_embed_model.load_model(file_name = 'bm_fit_model.pkl')

# setup dense embed model
dense_embed_model = LaBSEMultiLingualEmbedding()

# reranker
reranker_model = JinaAIMultiLingualReranker()

engine = RecommendationEngine(index_name, pinecone_apikey, sparse_embed_model, dense_embed_model, reranker_model)
```
