```python
import asyncio
from tqdm.asyncio import tqdm
import voyageai
import os
from pinecone import Pinecone


async def process_dataset(df, vo_client, _):
    batch_size = 128
    len_df = len(df)
    
    async for i in tqdm(range(0, len_df, batch_size)):
        i_end = min(i+batch_size, len_df)
        
        metadata_batch = df.iloc[i:i_end]['merged_metadata'].values.tolist()
        meta_dict = df.iloc[i:i_end]['metadata'].values.tolist()
        mergedtext_batch = df.iloc[i:i_end]['merged_text'].values.tolist()
        
        sparse_embeds = bm25.encode_documents([text for text in metadata_batch])
        dense_embeds = list(vo_client.embed(mergedtext_batch, model="voyage-multilingual-2", input_type="document", truncation=True).embeddings)
        product_ids = df.iloc[i:i_end]['product_id'].values.tolist()
        
        upserts = []
        for _id, sparse, dense, meta in zip(product_ids, sparse_embeds, dense_embeds, meta_dict):
            upserts.append({
                'id': _id,
                'sparse_values': sparse,
                'values': dense,
                'metadata': meta
            })
        
        await asyncio.to_thread(index.upsert, upserts)


async def main():
    # Create VoyageAI clients
    vo1 = voyageai.Client(api_key=os.environ.get('VOYAGE_AI1'))
    vo2 = voyageai.Client(api_key=os.environ.get('VOYAGE_AI2'))
    vo3 = voyageai.Client(api_key=os.environ.get('VOYAGE_AI3'))
    vo4 = voyageai.Client(api_key=os.environ.get('VOYAGE_AI4'))

    # Prepare datasets
    df_1= df.iloc[31104:106000,:] #df_1= df.iloc[:106000,:] Break
    df_2 = df.iloc[136976:212000,:]  #df_2 = df.iloc[106000:212000,:] 
    df_3 = df.iloc[242976:318000,:] # df_3 = df.iloc[212000:318000,:] 242976
    df_4 = df.iloc[348976:,:] # df_4 = df.iloc[318000:,:] 
    # notes : embedding model broke near iter : 242 ,manually identified positions


    # Create tasks for concurrent processing
    tasks = [
        process_dataset(df_1, vo1, "Dataset 1"),
        process_dataset(df_2, vo2, "Dataset 2"),
        process_dataset(df_3, vo3, "Dataset 3"),
        process_dataset(df_4, vo4, "Dataset 4")
    ]

    # Run all tasks concurrently
    await asyncio.gather(*tasks)

    # Print index stats
    print(index.describe_index_stats())

await main()
```

```python
def get_pretty_results(top_k_results , as_dataframe = False , verbose = True):
    from tabulate import tabulate
    def truncate(string, length):
        return string if len(string) <= length else string[:length-3] + '...'
    
    flattened_data = []
    for product in top_k_results.matches:
        flattened_product = {'id': product['id'] , **product['metadata'] }
        flattened_data.append(flattened_product)
    df = pd.DataFrame(flattened_data)
    column_order = ['product_id', 'brand', 'title', 'color', 'locale']
    df = df[column_order]
    if verbose :
        df['title'] = df['title'].apply(lambda x: truncate(x, 80))
        df['color'] = df['color'].apply(lambda x: truncate(x, 30))
        print(tabulate(df, headers='keys', tablefmt='grid', showindex=False))
    if as_dataframe:
        return df
    return
```

```python
get_pretty_results(result , as_dataframe= False )
```

```python
def hybrid_scale(dense, sparse, alpha = 0.5):
    """Hybrid vector scaling using a convex combination

    alpha * dense + (1 - alpha) * sparse

    Args:
        dense: Array of floats representing
        sparse: a dict of `indices` and `values`
        alpha: float between 0 and 1 where 0 == sparse only
               and 1 == dense only
    """
    if alpha < 0 or alpha > 1:
        raise ValueError("Alpha must be between 0 and 1")
    # scale sparse and dense vectors to create hybrid search vecs
    hsparse = {
        'indices': sparse['indices'],
        'values':  [v * (1 - alpha) for v in sparse['values']]
    }
    hdense = [v * alpha for v in dense]
    return hdense, hsparse
```

```python
question = "dark blue french connection jeans for men"

# scale sparse and dense vectors
hdense, hsparse = hybrid_scale(dense, sparse, alpha=0.5)
# search
result = index.query(
    top_k=15,
    vector=hdense,
    sparse_vector=hsparse,
    include_metadata=True
)
# used returned product ids to get images
get_pretty_results(result , as_dataframe= True )
```
